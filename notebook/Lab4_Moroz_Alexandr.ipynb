Downloading and reading dataframe

```python
import pandas as pd
lab4_df=pd.read_csv('https://data.cityofnewyork.us/api/views/5rq2-4hqu/rows.csv?accessType=DOWNLOAD')
```

```python
lab4_df.head()
```

Create reduced table

```python
columnOfInterest = ['spc_latin', 'spc_common','status', 'health', 'tree_dbh', 'stump_diam']
lab4_df_reduced = lab4_df[columnOfInterest]
lab4_df_reduced
```

Count of NaNs

```python
nullValue_counts=lab4_df_reduced.isnull().sum()
nullValue_counts
```

View NaNs only

```python
NaNsOnly=lab4_df_reduced.loc[(lab4_df_reduced['spc_latin'].isnull() ) | (lab4_df_reduced['spc_common'].isnull() ) | (lab4_df_reduced['health'].isnull() ) ]
NaNsOnly.head()
```

```python
pd.value_counts(NaNsOnly['status'])
```

```python
lab4_df_reduced_zerosFiltered = lab4_df_reduced.loc[(lab4_df_reduced['tree_dbh'] != 0) | (lab4_df_reduced['stump_diam'] != 0)]
```

```python
len(lab4_df_reduced_zerosFiltered)
```

```python
lab4_df_reduced_zerosFiltered_grouped = lab4_df_reduced_zerosFiltered.groupby(
    ['status', 'spc_latin', 'spc_common', 'health'], dropna=False
).mean()
```

```python
lab4_df_reduced_zerosFiltered_grouped
```

```python
lab4_df_reduced_zerosFiltered_grouped.reset_index()
```

```python
lab4_df_reduced_zerosFiltered_grouped['diameter'] = lab4_df_reduced_zerosFiltered_grouped['tree_dbh'] + lab4_df_reduced_zerosFiltered_grouped['stump_diam']
lab4_df_reduced_zerosFiltered_grouped
```

```python
lab4_df_reduced_zerosFiltered_grouped.drop(columns=['tree_dbh', 'stump_diam'], inplace=True)
lab4_df_reduced_zerosFiltered_grouped
```

```python
lab4_df_reduced_counts=lab4_df_reduced.assign(count = 1)
lab4_df_reduced_counts
```

```python
lab4_df_reduced_counts_grouped = lab4_df_reduced_counts.groupby(
    ['status', 'spc_latin', 'spc_common', 'health'], dropna=False
).sum().reset_index()
lab4_df_reduced_counts_grouped
```

```python
lab4_df_reduced_counts_grouped.drop(columns=['tree_dbh','stump_diam'], inplace=True)
lab4_df_reduced_counts_grouped=lab4_df_reduced_counts_grouped.reset_index()
lab4_df_reduced_counts_grouped.head()
```

```python
lab4_df_final = lab4_df_reduced_counts_grouped.merge(lab4_df_reduced_zerosFiltered_grouped,
                                                     on=['status', 'spc_latin', 'spc_common', 'health'],
                                                     how='left')
lab4_df_final
```

```python
lab4_df_final.drop(columns=['index'], inplace=True)
lab4_df_final
```

```python
lab4_df_final=lab4_df_final.set_index(['status','spc_latin','spc_common','health'])
lab4_df_final
```

```python
lab4_df_final.to_csv('lab1_df_final.csv')
```
